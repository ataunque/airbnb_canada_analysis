```python
import pandas as pd
from collections import Counter
from nltk.corpus import stopwords
from stop_words import get_stop_words
```

```python
#Stop Words Intialise
pd.set_option('display.max_rows',100)
my_stopwords =[
    'like',
    'one',
    'much',
    'even',
    'also',
    'get',
    'made',
    'pour',
    'thanks',
    'right',
    'looking',
    'sure',
    'come',
    'feel',
    'anyone',
    'find',
    'ever',
    'always',
    'things',
    'thought',
    'plus',
    'lot',
    'use',
    'dans',
    'time',
    'day',
    'minutes',
    'way',
    'minute',
    'days',
    'people',
    'need',
    'lots',
    'check',
    'back'
]

stopWords = get_stop_words('english')+stopwords.words('english')+get_stop_words('french')+stopwords.words('french')+my_stopwords
```

```python
listings = pd.read_csv('Data\great_listings.csv')

listings.head()
```

```python
#We are going to look at the most offered amenities amongst our top listings. 
# The data is a strings so we remove all non alpahbet character, except the spaces and commas
listings['amenities'] = listings['amenities'].str.replace(r'[^a-zA-Z, ]', '', regex=True)

#We then create a combined string of all ameniities from every listing and use the .split() method to create a list of the words, split on the comma. 
amenities = listings['amenities'].str.cat()
amenities = amenities.split(',')

#Counter then counts the occurances
amenities_count = Counter(amenities)

#Convert the counter dictionary into a pandas dataframe. We make it a list first and then sort it in descending to get the top used words. 
amenities_count = pd.DataFrame(list(amenities_count.items()), columns=['Amenity', 'Count'])
amenities_count=amenities_count.sort_values(by='Count',ascending=False)

amenities_count.head(10)
```

```python
#Let's see what the most common words are used by guest when reviewing these great listings
listings['reviews'] = listings['reviews'].str.replace(r'.<br/>', '', regex=True)
listings['reviews'] = listings['reviews'].str.replace(r'[^a-zA-Z ]', ' ', regex=True)
listings['reviews']
#We then create a combined string of all reviews from every listing and use the .split() method to create a list of the words, split on the comma. 
reviews = listings['reviews'].str.cat()
reviews = reviews.split(' ')

reviews_cleaned= []

for word in reviews:
    if word.lower() not in stopWords and len(word) > 2 :
        reviews_cleaned.append(word.lower())

#Counter then counts the occurances
reviews_count = Counter(reviews_cleaned)

#Convert the counter dictionary into a pandas dataframe. We make it a list first and then sort it in descending to get the top used words. 
reviews_count = pd.DataFrame(list(reviews_count.items()), columns=['Review_Term', 'Count'])
reviews_count=reviews_count.sort_values(by='Count',ascending=False)

reviews_count.head(10)
```

```python
#export files
amenities_count.to_csv('Data\\amenities_count.csv',index=False)
reviews_count.to_csv('Data\\reviews_count.csv',index=False)
```
